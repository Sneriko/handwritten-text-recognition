```python
batch_name = 'B0001097'
batch_path = '../../../data/1930_census/Gotland/B0001097'
```

```python
db_gotland = pd.read_csv('../../../data/1930_census/Gotland/Folkräkning_1930_Gotlands_län', '\t', dtype=str)

batch_filter = pd.read_excel('../../../data/1930_census/Gotland/B0001097/B0001097_filter.xlsx', dtype=str)
```

```python
#fixing asynchronicity lists btw special and stand

p_stand = os.path.join(batch_path, 'B0001097_standard')
p_spec = os.path.join(batch_path, 'B0001097_crop_special')

stand = glob(p_stand + '/**')
spec = glob(p_spec + '/**')

stand = [x.split('/')[-1] for x in stand]
spec = [x.split('/')[-1] for x in spec]

stand.sort()
spec.sort()

use_stand = []
use_spec = []

for p in spec:
    for i, x in enumerate(stand):
        if p == x:
            break
        elif i == len(stand) - 1:
            use_spec.append(p)

for p in stand:
    for i, x in enumerate(spec):
        if p == x:
            break
        elif i == len(spec) - 1:
            use_stand.append(p)

pages = list(set(spec + stand))

for i, page in enumerate(pages):

    #print(str(i))
    
    if page in use_spec or page in batch_filter.spec_year.values:
        write_page_imgs(crop = 'crop_special',
                        name = page,
                        db = db_gotland,
                        batch_filter = batch_filter,
                        batch_path = batch_path)

    elif page in use_stand or page in batch_filter.stand_year.values:
        write_page_imgs(crop = 'standard',
                        name = page,
                        db = db_gotland,
                        batch_filter = batch_filter,
                        batch_path = batch_path)

    else:
        coin = random.randint(0, 1)
        
        if coin == 0:
            write_page_imgs(crop = 'crop_special',
                            name = page,
                            db = db_gotland,
                            batch_filter = batch_filter,
                            batch_path = batch_path)

        else:
            write_page_imgs(crop = 'standard',
                            name = page,
                            db = db_gotland,
                            batch_filter = batch_filter,
                            batch_path = batch_path)


```

```python
import glob
from pathlib import Path 

imgs = glob.glob('/content/drive/MyDrive/Riksarkivet/htr/data/CycleGAN_Datasets/only_overwritten/testA/*.jpg')
imgs.sort()


for i, im in enumerate(imgs):
  img = cv2.imread(im)
  
  dst = cv2.fastNlMeansDenoising(img, h=31, templateWindowSize=7, searchWindowSize=21)
  img_grey = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY)
  img_blur = cv2.medianBlur(img_grey,3).astype('uint8')
  img_thresh_Gaussian = cv2.adaptiveThreshold(img_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

  cv2.imwrite('/content/drive/MyDrive/Riksarkivet/htr/CycleGAN_models/Only_overwritten/only_cv2/test_%s.jpg' % str(i).zfill(2), img_thresh_Gaussian)
```
